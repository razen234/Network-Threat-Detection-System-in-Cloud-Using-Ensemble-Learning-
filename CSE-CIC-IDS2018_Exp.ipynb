```python
import os
import pandas as pd
import numpy as np
```

```python
dataset = "CSE-CIC-IDS2018"
```

```python
import os
print("Current working directory:", os.getcwd())
print("Contents:", os.listdir())
```

```python
csv_files = [f for f in os.listdir(dataset) if f.endswith(".csv")]
csv_files.sort()
```

```python
print("CSV files inside", dataset, ":")
for f in csv_files:
    print(f)
```

```python
for f in csv_files:
    path = os.path.join(dataset, f)
    df = pd.read_csv(path, low_memory=False, on_bad_lines="skip")
    # Drop stray "Label" rows if they exist
    if "Label" in df.columns:
        df = df[df["Label"] != "Label"]
    print(f"\n=== {f} ===")
    print(df["Label"].value_counts().head(10))  #
```

```python
attack_files = [
    "02-14-2018.csv", 
    "02-15-2018.csv",  
    "02-16-2018.csv", 
    "02-21-2018.csv", 
    "02-22-2018.csv", 
    "02-23-2018.csv",  
    "02-28-2018.csv",  
    "03-01-2018.csv",
    "03-02-2018.csv", 
]
```

```python
dfs = []
for fname in attack_files:
    path = os.path.join(dataset, fname)
    print("Loading:", path)
    df = pd.read_csv(path, low_memory=False, on_bad_lines="skip")
    if "Label" in df.columns:
        df = df[df["Label"] != "Label"]
    dfs.append(df)
```

```python
data = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
data.to_csv("CIC-IDS-2018-attacks.csv", index=False)
print("Saved merged attack dataset -> CIC-IDS-2018-attacks.csv")
print("Final shape:", data.shape)
print("Label distribution:\n", data["Label"].value_counts())
```
